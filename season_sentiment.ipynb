```python
import matplotlib.pyplot as plt
import pandas as pd

book_emotions_df = pd.read_csv('book_level_emotions.csv')

book_emotions_df.set_index('book').plot(kind='bar', figsize=(12, 6), stacked=False)
plt.title("Emotion Comparison by Book")
plt.ylabel("Emotion Count")
plt.xlabel("Book")
plt.legend(title="Emotions", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```
